```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
try:
    df = pd.read_csv('owid-covid-data.csv')
except FileNotFoundError:
    print("Error: 'owid-covid-data.csv' not found.")
    exit()

# Filter data
countries = ['South Africa', 'United States', 'India']
df_filtered = df[df['location'].isin(countries) & (df['date'] >= '2020-01-01') & (df['date'] <= '2022-12-31')].copy()
df_filtered['date'] = pd.to_datetime(df_filtered['date'])

# Function to create line plots
def plot_data(df, x, y, title, y_label):
    plt.figure(figsize=(10, 5))
    sns.lineplot(data=df, x=x, y=y, hue='location')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel(y_label)
    plt.legend(title='Country')
    plt.grid(True)
    plt.show()

# 3. Plot Total Cases Over Time
plot_data(df_filtered, 'date', 'total_cases', 'Total COVID-19 Cases Over Time', 'Total Cases')

# 4. Plot Total Deaths Over Time
plot_data(df_filtered, 'date', 'total_deaths', 'Total COVID-19 Deaths Over Time', 'Total Deaths')

# 5. Compare Daily New Cases Between Countries
plot_data(df_filtered, 'date', 'new_cases', 'Daily New COVID-19 Cases', 'Daily New Cases')

# 6. Calculate and Plot Death Rate
df_filtered['death_rate'] = (df_filtered['total_deaths'] / df_filtered['total_cases']) * 100
plot_data(df_filtered, 'date', 'death_rate', 'COVID-19 Death Rate (%)', 'Death Rate (%)')

# Smoother trend for daily new cases (optional)
def plot_rolling_average(df, column='new_cases', window=7):
    plt.figure(figsize=(10, 5))
    for country in countries:
        country_data = df[df['location'] == country].copy()
        country_data['rolling_avg'] = country_data[column].rolling(window=window).mean()
        plt.plot(country_data['date'], country_data['rolling_avg'], label=country)
    plt.title(f'Rolling Average ({window}-day) of Daily New COVID-19 Cases')
    plt.xlabel('Date')
    plt.ylabel('Rolling Average of New Cases')
    plt.legend(title='Country')
    plt.grid(True)
    plt.show()

plot_rolling_average(df_filtered)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('owid-covid-data.csv')

# Filter data
countries = ['South Africa', 'United States', 'India']
vaccine_data = df[df['location'].isin(countries) & (df['date'] >= '2020-01-01')].copy()
vaccine_data['date'] = pd.to_datetime(vaccine_data['date'])

# Plot cumulative vaccinations
plt.figure(figsize=(10, 5))
sns.lineplot(data=vaccine_data.dropna(subset=['total_vaccinations']), x='date', y='total_vaccinations', hue='location')
plt.title('Cumulative Vaccinations')
plt.xlabel('Date')
plt.ylabel('Total Vaccinations')
plt.yscale('log')
plt.legend(title='Country')
plt.grid(True)
plt.show()

# Plot % vaccinated
plt.figure(figsize=(10, 5))
sns.lineplot(data=vaccine_data.dropna(subset=['people_vaccinated_per_hundred']), x='date', y='people_vaccinated_per_hundred', hue='location')
plt.title('% Vaccinated Population')
plt.xlabel('Date')
plt.ylabel('% Vaccinated')
plt.legend(title='Country')
plt.grid(True)
plt.show()
```
